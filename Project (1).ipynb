```python
df.drop_duplicates(inplace=True)
```

```python
df['education_level'].unique()
df.columns
```

```python
data = df.drop(['height_cm', 'weight_kg', 'num_siblings', 'has_pet',
                'favorite_color', 'birth_month'], axis=1)
data
```

```python
data['preferred_device'].unique()
```

```python
dummies = pd.get_dummies(data.preferred_device,drop_first=True).astype(int)
dummies
```

```python
merged = pd.concat([data,dummies], axis=1)
merged
```

```python
#from sklearn.preprocessing import OneHotEncoder
#ohe=OneHotEncoder(sparse_output=False)
```

```python
#X = ohe.fit_transform(data[['preferred_device']])
#print(X[:5])
```

```python
X=merged.drop(['continent', 'country', 'education_level', 'preferred_device','completed_course'],axis=1)
X
```

```python
y=merged['completed_course']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression(max_iter=1000)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
y_predict = reg.predict(X_test)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
plt.figure()
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
plt.figure()
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_predict))
```
